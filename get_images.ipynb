```python
import os
import requests
from pygbif import occurrences
```

```python
CEBOLINHA_SPECIES_ID = 2855860
LIMIT = 500
```

```python
fields = ['key', 'taxonKey', 'datasetKey', 'basisOfRecord' 'continent', 'country', 'media']
occur_cebolinha = occurrences.search(taxonKey = CEBOLINHA_SPECIES_ID, mediatype='StillImage', limit=LIMIT)

img_cebolinha = [
    media_item['identifier']
    for result in occur_cebolinha['results']
    if 'media' in result
    for media_item in result['media']
    if 'identifier' in media_item
]
```

```python
def download_image(img_list, species):
    folder = f'images\{species}'
    os.makedirs(folder, exist_ok=True)
    count = 0
    
    for url in img_list:
        try:
            response = requests.get(url, stream=True)
            response.raise_for_status()
            fileextension = url.split(".")[-1]
            filename = os.path.join(folder, f'{count}.{fileextension}')
            with open(filename, 'wb') as out_file:
                for chunk in response.iter_content(chunk_size=8192):
                    out_file.write(chunk)
            count += 1
            print(f'Imagem salva como: {filename}')
        except requests.exceptions.RequestException as e:
            print(f'Erro ao baixar a imagem de {url}: {e}')
```

```python
download_image(img_cebolinha, 'allium-schoenoprasum')
```
