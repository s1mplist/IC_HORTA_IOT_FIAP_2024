```python
import cv2
import numpy as np
import tensorflow as tf
```

```python
#Load the saved model
model = tf.keras.models.load_model('plantVillage.keras')
video = cv2.VideoCapture(0)

# Parâmetros do modelo (adapte se necessário)
img_width, img_height = 224, 224  # Ajustar de acordo com o treinamento
num_classes = 39  # Ajustar de acordo com o número de classes
```

```python
# Função para pré-processar a imagem
def preprocess_image(img):
    # Redimensionar a imagem
    img = cv2.resize(img, (img_width, img_height))
    # Normalizar os valores dos pixels
    img = img / 255.0
    # Adicionar um eixo para representar o batch
    img = np.expand_dims(img, axis=0)
    return img
```

```python
# Capturar vídeo da webcam
cap = cv2.VideoCapture(0)

while True:
    # Ler um frame da webcam
    ret, frame = cap.read()

    # Pré-processar a imagem
    img_preprocessed = preprocess_image(frame)

    # Fazer a predição
    predictions = model.predict(img_preprocessed)
    predicted_class = np.argmax(predictions)

    # Exibir a classe predita
    cv2.putText(frame, f"Classe: {predicted_class}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Exibir o frame
    cv2.imshow('Video', frame)

    # Sair do loop se a tecla 'q' for pressionada
    if cv2.waitKey(1) == ord('q'):
        break

# Liberar a câmera e fechar as janelas
cap.release()
cv2.destroyAllWindows()
```
